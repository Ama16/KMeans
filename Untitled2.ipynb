```python
import numpy as np
import random
import matplotlib
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import time
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform
from sklearn.preprocessing import normalize
```

```python
random.seed(16)
```

```python
class MyKMeans:
    def __init__(self, k=2, metric='euclidean'):
        self.k = k
        
        if metric == 'euclidean':
            self.get_dist = self.distance_euclidean
        elif metric == 'manhattan':
            self.get_dist = self.distance_manhattan
    
    def distance_euclidean(self, x, y):
        return (np.sum(np.sqrt((x-y)**2)))
    
    def distance_manhattan(self, x, y):
        return np.sum(np.abs(x-y))
    
    def calculate_color(self):
        dist = np.empty(self.k);
        for i in range(len(self.X)):
            for j in range(self.k):
                dist[j] = self.get_dist(self.X[i], self.center[j])
            self.color[i] = np.argmin(dist)
            
            
    def calculate_center(self):
        for i in range(self.k):
            need = self.color == i
            summ = np.zeros(self.X.shape[1])
            for j in self.X[need]:
                summ += j
            summ /= len(self.X[need])
            self.center[i] = summ
            
        
    def fit_predict(self, X):
        self.X = X
        
        self.center = np.empty((self.k, X.shape[1]))
        #for i in range(self.k):
        #  self.center[i] = self.X[i]
        
        self.center[0] = self.X[0]
        
        for i in range(1, self.k):
            summ = 0
            for j in range(len(self.X)):
                minim = -1
                for k in range(i):
                    if self.get_dist(self.center[k], self.X[j]) ** 2 < minim or minim == -1:
                        minim = self.get_dist(self.center[k], self.X[j]) ** 2
                        
                summ += minim
            rnd = random.random() * summ
            summ = 0
            j = 0
            while summ <= rnd:
                minim = -1
                for k in range(i):
                    if self.get_dist(self.center[k], self.X[j]) ** 2 < minim or minim == -1:
                        minim = self.get_dist(self.center[k], self.X[j]) ** 2
                        
                summ += minim
                j += 1
            j -= 1
            self.center[i] = self.X[j]
        
        
            
        self.color = np.ones(len(self.X))
        
        a = np.zeros(len(self.X))
        old_center = np.zeros((self.k, X.shape[1]))
        while np.any(old_center-self.center) > 0:
            old_center = self.center
            a = self.color
            self.calculate_color()
            self.calculate_center()
        self.calculate_color()
        return self.color
        
    def print(self):
        return self.center
    
    def dist_sum(self):
        sum = 0
        for i in range(len(self.X)):
            sum += self.get_dist(self.X[i], self.center[int(self.color[i])])
        return sum
    
    def plot_dist_matrix(self):
        dist = pdist(self.X)
        X_sort = np.empty(self.X.shape[1])
        for i in range(len(np.unique(self.color))):
            for k, j in enumerate(self.X):
                if (self.color[k] == i):
                    X_sort = np.vstack((X_sort, j))
                    
        norm = normalize(squareform(pdist(X_sort[1:])), axis=1, norm='l1')
        plt.matshow(norm, fignum=1, cmap=plt.get_cmap('Blues_r'))
        plt.colorbar()
        
    
```

```python
x_min, x_max = blob[:, 0].min() - .5, blob[:, 0].max() + .5
y_min, y_max = blob[:, 1].min() - .5, blob[:, 1].max() + .5

plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(blob[:, 0], blob[:, 1], c=blob_labels)

plt.xlabel('Y')
plt.ylabel('X')

model = MyKMeans(4)
model.fit_predict(blob)
x, y = model.print().T

plt.scatter(x, y, c = 'red', s = 100)

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())
```

Теперь посмотрим, как выбирать гиперпараметр k. Можно выбирать его так называемым методом локтя. Смотрим на зависимость суммарного расстояния от центров до элементов, которые при классификации попали в класс этого центра, при различном значении параметра k; как только расстояние перестанет значительно уменьшаться при увеличении k, фиксируем это значение и полагаем, что оно оптимально

Попробуем посмотреть, что получится с данными, которые мы уже использовали

```python
centers = [[1, 0], [-1, -1], [1, -2], [2, 2]]
blob, _ = make_blobs(n_samples=1500, n_features=2, centers=centers, random_state=16, cluster_std=0.2)
```

```python
distance = []

for i in range(2, 10):
    model = MyKMeans(k=i)
    model.fit_predict(blob)
    distance.append(model.dist_sum())

plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.plot(range(2, 10), distance, marker='s');
plt.xlabel('K')
plt.ylabel('Distance');

plt.subplot(1, 2, 2)
model = MyKMeans(k=4)

plt.scatter(blob[:, 0], blob[:, 1], c=model.fit_predict(blob))
```

Видим, что на значении k=4 сумма расстояний слабо изменяется, и значение k=4 реально является оптимальным

```python
centers = [[1, 0], [-1, 0], [2, 1]]
blob, _ = make_blobs(n_samples=1500, n_features=2, centers=centers, random_state=16, cluster_std=0.4)
```

```python
distance = []

for i in range(2, 10):
    model = MyKMeans(k=i)
    model.fit_predict(blob)
    distance.append(model.dist_sum())

plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.plot(range(2, 10), distance, marker='s');
plt.xlabel('K')
plt.ylabel('Distance');

plt.subplot(1, 2, 2)
model = MyKMeans(k=3)

plt.scatter(blob[:, 0], blob[:, 1], c=model.fit_predict(blob))
```
